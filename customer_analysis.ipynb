# So we fill null values with median but for more accuracy we will fill the null values with median based on category.

# i will be using groupby to group category wise and then transform the column review rating with fillna using lambda function 

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

# Now the data is cleaned as there is no null values in any column 

# now we can se the capital letter and spaces within the column we will change it to lower cases with _ as we will be using them in every line

```python
df.columns= df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
```

```python
df.columns
```

# here we can see that one column purchase amount is left to be changed so will rename it

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

# here we can see that all the columns are now perfectly transformed

## Now we will make derived columns 

## we will now make age group based on age from the age column

```python
labels =['Young adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df.columns
```

```python
df[['age','age_group']].head(20)
```

## Now we will see that how frequently the order is being placed like monthly daily or weekly or quarterly

### for this we will use dictionary to map


```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(20)
```

## Now we will se that the discount which is applied is due to promo code or something else

## We can now drop promo code column as we can see that all discount is due to promo code


```python
df = df.drop('promo_code_used',axis=1)
```
